```python
destination="C:\\Users\\sayan\\Downloads"
f=os.listdir(destination)
os.chdir(destination)
```

```python
#getting the name of all assignment files
f_specific=[]
for i in f:
    if 'Assingment' in i:
        f_specific.append(i)
        
for i in f:
    if 'Assignment' in i:
        f_specific.append(i)    
f_specific
```

```python
import re
```

```python
#creating name specific folders
f_folder=[]
for i in f_specific:
    r=re.search('\d',i)   #searching the 1rst occurrence of digit
    f_folder.append(i[:r.start()].replace("_"," "))   #appending the value before the frst occurance of digit
    if not os.path.isdir(i[:r.start()].replace("_"," ")):
        os.mkdir(i[:r.start()].replace("_"," "))     #creating the files
f_folder
```

```python
#changing the file names
f_rename=[]
for i in f_specific:
    if "(" and ")" in i:
        temp=i.replace("(","")
        temp=temp.replace(" ","_")
        f_rename.append(temp.replace(")",""))
    else:
        f_rename.append(i) 
f_rename
```

```python
#renaming the actual file
for i in range(len(f_rename)):
    os.rename(f_specific[i],f_rename[i])
```

```python
#creating the master docs where all the question exist
import docx
from docx import Document
doc = Document()
for i in f_rename:
    doc.add_heading(i,2)
    doc1 = Document(i)
    for para in doc1.paragraphs:
        doc.add_paragraph(para.text)
doc.save("master_doc.docx")
```

```python
#moving the file in respective directory
import shutil
```

```python
Programming_Assingment="C:\\Users\\sayan\\Downloads\\Programming Assingment"
Assignment="C:\\Users\\sayan\\Downloads\\Assignment"
for i in f_rename:
    if "Programming_Assingment" in i:
        shutil.move(destination+"\\"+i, Programming_Assingment )
    if "Assignment" in i:
        shutil.move(destination+"\\"+i, Assignment )
```
